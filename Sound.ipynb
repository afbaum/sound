### Lets make a sin wave

We have to import extra libraries into Python.  This gives us access to extra functions making it eaier to write our code

We need to collect a sample of data.  A basic sin wave is nothing more than calculating sin based on a period of time.  In the following code we create an array that has a bunch of different time intervals ranging from 0 to 4pi in steps of 0.1.  

```python
import numpy as np
import scipy
import pandas as pd
import matplotlib.pyplot as plot

t = np.arange(0, 4*np.pi, 0.1)
u = np.linspace(0, 4*np.pi, 500)
```

Next we are going to calculate and plot the sin of each of the numbers listed in the array above.

```python

s = np.sin(t)
z = np.sin(u)

plt.ylabel("Amplitude")
plt.xlabel("Time [s]")
plt.title('Plot of sin wave from 0 to 4pi')
plt.plot(t, s)
plt.show()

plt.ylabel("Amplitude")
plt.xlabel("Time [z]")
plt.title('Plot of sin wave from 0 to 4pi')
plt.plot(u, z)
plt.show()
```

The numpy arange function allows us to create an array with the starting number, ending number 
and the steps size used to get from the start to the end.

## How do we change the amplitude and frequency of the sin wave?

The actual formula for a sin wave is 
    y(t) = A * sin(wt + deg)
    
    A is the amplitude of the wave
    w = the angular frequency
    deg = the phase

```python
ts = np.linspace(0, 1, 500)
newSin = 10 * np.sin(50 * ts)

nextSin = 5 * np.sin(220 *ts)

lastSin = 7.5 *  np.sin(110 * ts)

plt.ylabel("Amplitude")
plt.xlabel("Time [newSin]")
plt.title('Plot of sin wave from 0 to 4pi')
plt.plot(ts, newSin, ts, nextSin, ts, lastSin)
plt.show()
```
